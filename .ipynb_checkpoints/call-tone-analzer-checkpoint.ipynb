### Detect language of review and select only 2000 rows


```python
en_reviews= []
en_count=0

from langdetect import detect
for index, review in enumerate(hotels_reviews['reviews.text']):
    try:
        print(detect(review))
        if detect(review) == 'en':
            en_reviews.append(index)
            print(index)
            en_count+= 1 
            #print(en_count)
            
            #print("other lang>>>",index)
    except:
        print('no lang Detected')
    if en_count== 2000:
        break
```

```python
hotels_reviews= hotels_reviews.iloc[en_reviews]
```

## call tone analyzer API and retrive tones

```python
from ibm_watson import ToneAnalyzerV3
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator

authenticator = IAMAuthenticator('Fpv6KvXVlLfEP67Jx33Y9LbbhBJv_ep66ubRNNYL1kq2')
tone_analyzer = ToneAnalyzerV3(
    version='2017-09-21',
    authenticator=authenticator )
tone_analyzer.tone
tone_analyzer.set_service_url('https://gateway-lon.watsonplatform.net/tone-analyzer/api')
```

```python
# select only document tone from tone analyzer response
import json
def analyze_text(text):
    tone_analysis = tone_analyzer.tone(
        {'text': text},
        content_type='application/json'
        
    ).get_result()
    response= json.dumps(tone_analysis, indent=2)
    
    tones= json.loads(response)
    if tones['document_tone']['tones']==[]:
        return []
    
    return tones['document_tone']['tones']
    #return response
```
