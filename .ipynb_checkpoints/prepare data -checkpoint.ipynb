```python
columns_list=['categories' ,'reviews.date', 'reviews.dateAdded',
       'reviews.doRecommend', 'reviews.id','reviews.title','reviews.userCity', 'reviews.username',
       'reviews.userProvince']
```

```python
hotel_data= hotel_data.drop(columns_list,axis=1)
```

```python
hotel_data.to_csv('hotel_data.csv')
```

```python
import json
import numpy as np
```

```python
#gropby name and keep data['address','city','country', 'province'] of hotel in dataframe.
```

```python
df2 = pd.DataFrame(columns=['address','city','country', 'province'])
```

```python
df2['address'] = hotel_data.groupby('name')['address'].apply(lambda x: np.unique(x)[0])
```

```python
df2['city'] = hotel_data.groupby('name')['city'].apply(lambda x: np.unique(x)[0])
```

```python
df2['country']= hotel_data.groupby('name')['country'].apply(lambda x: np.unique(x)[0])
```

```python
df2['province']= hotel_data.groupby('name')['province'].apply(lambda x: np.unique(x)[0])
```

```python
df2
```

```python
hotel_data.groupby('name')['address','city','country', 'province'].apply(lambda x: [np.unique(x)[0], np.unique(x)[1],np.unique(x)[2],np.unique(x)[3]])
```

```python
final_data_frame =hotel_data.groupby('name').mean()
```

```python
hotels_db= pd.concat([final_data_frame, df2], axis=1)
```

```python
# replace nan values with zeros
hotels_db= hotels_db.fillna(0)
```

```python
hotels_db
```

```python
hotels_db = hotels_db.reset_index()
```

```python
# saving data into list of dicts to use it later in elasticsearch
import json 
hotel_json =hotels_db.to_json(orient='records')
doc= json.loads(hotel_json)
```
