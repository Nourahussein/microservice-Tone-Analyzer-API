```python
import json
def analyze_text(text):
    tone_analysis = tone_analyzer.tone(
        {'text': text},
        content_type='application/json'
        
    ).get_result()
    response= json.dumps(tone_analysis, indent=2)
    
    tones= json.loads(response)
    if tones['document_tone']['tones']==[]:
        return []
    
    return tones['document_tone']['tones']
    #return response
```
